## **🟢 Part 1: Loan Eligibility & Interest Rate Calculation [4 marks]**  

📌 **Requirement:**  
SecureBank only approves loans if the **customer meets all eligibility criteria**. The system should:  
✔ **Check if the customer is employed.**  
✔ **Verify if the income is at least PKR 50,000.**  
✔ **Check credit score:**  
   - **750+ → 5% Interest Rate**  
   - **650 - 749 → 8% Interest Rate**  
   - **Below 650 → Loan Rejected**  
✔ **If the applicant is unemployed, the loan is rejected immediately.**  

### **Implementation Steps:**  
🔹 **Step 1:** Structure the `if-elif-else` statements for this logic.  
🔹 **Step 2:** Implement a Python program using **ipywidgets** for interactive user input (income, credit score, employment status) to decide whether the loan is approved or rejected.

---

```python
import ipywidgets as widgets
from IPython.display import display

def check_loan_eligibility(employed, income, credit_score):
    if not employed:
        return "Loan Rejected: Applicant is unemployed."
    if income < 50000:
        return "Loan Rejected: Income below PKR 50,000."
    if credit_score >= 750:
        return "Loan Approved: Interest Rate 5%."
    elif 650 <= credit_score < 750:
        return "Loan Approved: Interest Rate 8%."
    else:
        return "Loan Rejected: Credit score below 650."

def on_submit(_):
    result = check_loan_eligibility(employment_checkbox.value, income_slider.value, credit_score_slider.value)
    output_label.value = result

employment_checkbox = widgets.Checkbox(description="Employed", value=False)
income_slider = widgets.IntSlider(description="Income", min=0, max=200000, step=5000, value=50000)
credit_score_slider = widgets.IntSlider(description="Credit Score", min=300, max=850, step=10, value=700)
submit_button = widgets.Button(description="Check Loan Eligibility")
output_label = widgets.Label()

submit_button.on_click(on_submit)

display(employment_checkbox, income_slider, credit_score_slider, submit_button, output_label)
```

## **🟢 Part 3: Loan Repayment Tracker [4 marks]**  

📌 **Requirement:**  
Customers who receive a loan should be able to track their **loan balance** as they make monthly payments. The system should:  
✔ Start with an **initial loan balance** (e.g., PKR 500,000).  
✔ Deduct a **fixed monthly payment** (e.g., PKR 25,000).  
✔ Continue tracking until **loan balance reaches zero**.  
✔ Display the remaining balance **after each payment**.  

### **Implementation Steps:**  
🔹 **Step 1:** Choose an appropriate loop (`for` or `while`).  
🔹 **Step 2:** Ensure the loop **stops once the loan is fully paid**.  
🔹 **Step 3:** Implement a Python program using **ipywidgets** to **simulate loan repayment interactively**.

---

```python
import ipywidgets as widgets
from IPython.display import display

def classify_risk(stock_returns):
    if any(r < 0 for r in stock_returns):
        return "High Risk: At least one stock has a negative return."
    elif all(r >= 5 for r in stock_returns):
        return "Low Risk: All stock returns are 5% or above."
    else:
        return "Medium Risk: All returns are positive, but at least one is below 5%."

def track_loan_repayment(initial_balance, monthly_payment):
    balance = initial_balance
    payments = []
    while balance > 0:
        balance -= monthly_payment
        balance = max(balance, 0)
        payments.append(balance)
    return payments

def on_submit(_):
    returns = [float(val) for val in stock_returns_box.value.split(',')]
    result = classify_risk(returns)
    output_label.value = result

def on_loan_submit(_):
    initial_balance = float(initial_balance_box.value)
    monthly_payment = float(monthly_payment_box.value)
    payments = track_loan_repayment(initial_balance, monthly_payment)
    loan_output_label.value = "Remaining Balances: " + str(payments)

stock_returns_box = widgets.Text(description="Stock Returns", placeholder="Enter returns, e.g., 3, 7, -2")
submit_button = widgets.Button(description="Evaluate Risk")
output_label = widgets.Label()
submit_button.on_click(on_submit)

display(stock_returns_box, submit_button, output_label)

initial_balance_box = widgets.Text(description="Initial Loan", placeholder="Enter initial balance, e.g., 500000")
monthly_payment_box = widgets.Text(description="Monthly Payment", placeholder="Enter monthly payment, e.g., 25000")
loan_submit_button = widgets.Button(description="Track Loan Repayment")
loan_output_label = widgets.Label()
loan_submit_button.on_click(on_loan_submit)

display(initial_balance_box, monthly_payment_box, loan_submit_button, loan_output_label)
```


## **🟢 Part 4: Stock Price Monitoring and Trading Strategy [4 marks]**  

📌 **Requirement:**  
A stock trader wants to **track stock prices daily** and **sell when the price reaches PKR 200**. The system should:  
✔ Iterate through a **list of stock prices**.  
✔ **Skip missing stock data** (`None` values).  
✔ Stop tracking **once the price reaches PKR 200**.  

### **Implementation Steps:**  
🔹 **Step 1:** Handle missing stock data using `continue`.  
🔹 **Step 2:** Stop tracking once the stock hits the target price (`break`).  
🔹 **Step 3:** Write a Python program using **ipywidgets** to **process stock prices interactively** and trigger alerts when conditions are met.

---

```python
import ipywidgets as widgets
from IPython.display import display

def classify_risk(stock_returns):
    if any(r < 0 for r in stock_returns):
        return "High Risk: At least one stock has a negative return."
    elif all(r >= 5 for r in stock_returns):
        return "Low Risk: All stock returns are 5% or above."
    else:
        return "Medium Risk: All returns are positive, but at least one is below 5%."

def track_loan_repayment(initial_balance, monthly_payment):
    balance = initial_balance
    payments = []
    while balance > 0:
        balance -= monthly_payment
        balance = max(balance, 0)
        payments.append(balance)
    return payments

def track_stock_prices(stock_prices):
    tracked_prices = []
    for price in stock_prices:
        if price is None:
            continue
        if price >= 200:
            tracked_prices.append(price)
            break
        tracked_prices.append(price)
    return tracked_prices

def on_submit(_):
    returns = [float(val) for val in stock_returns_box.value.split(',')]
    result = classify_risk(returns)
    output_label.value = result

def on_loan_submit(_):
    initial_balance = float(initial_balance_box.value)
    monthly_payment = float(monthly_payment_box.value)
    payments = track_loan_repayment(initial_balance, monthly_payment)
    loan_output_label.value = "Remaining Balances: " + str(payments)

def on_stock_submit(_):
    stock_prices = [float(val) if val.lower() != 'none' else None for val in stock_prices_box.value.split(',')]
    tracked = track_stock_prices(stock_prices)
    stock_output_label.value = "Tracked Stock Prices: " + str(tracked)

stock_returns_box = widgets.Text(description="Stock Returns", placeholder="Enter returns, e.g., 3, 7, -2")
submit_button = widgets.Button(description="Evaluate Risk")
output_label = widgets.Label()
submit_button.on_click(on_submit)

display(stock_returns_box, submit_button, output_label)

initial_balance_box = widgets.Text(description="Initial Loan", placeholder="Enter initial balance, e.g., 500000")
monthly_payment_box = widgets.Text(description="Monthly Payment", placeholder="Enter monthly payment, e.g., 25000")
loan_submit_button = widgets.Button(description="Track Loan Repayment")
loan_output_label = widgets.Label()
loan_submit_button.on_click(on_loan_submit)

display(initial_balance_box, monthly_payment_box, loan_submit_button, loan_output_label)

stock_prices_box = widgets.Text(description="Stock Prices", placeholder="Enter prices, e.g., 180, 190, None, 205")
stock_submit_button = widgets.Button(description="Track Stock Prices")
stock_output_label = widgets.Label()
stock_submit_button.on_click(on_stock_submit)

display(stock_prices_box, stock_submit_button, stock_output_label)
```


## **🟢 Part 5: Currency Exchange Rate Tracker [4 marks]**  

📌 **Requirement:**  
SecureBank provides **real-time currency exchange tracking**. The system should:  
✔ Start at **PKR 290/USD**.  
✔ Increase by **1 PKR per day** until it reaches **PKR 300/USD**.  
✔ Print exchange rates daily and stop when the **target rate is reached**.  

### **Implementation Steps:**  
🔹 **Step 1:** Choose a suitable loop (`for` or `while`).  
🔹 **Step 2:** Stop the loop when the exchange rate reaches the target.  
🔹 **Step 3:** Implement a Python program using **ipywidgets** to **track the currency exchange rate interactively**.

---

```python
import ipywidgets as widgets
from IPython.display import display

def classify_risk(stock_returns):
    if any(r < 0 for r in stock_returns):
        return "High Risk: At least one stock has a negative return."
    elif all(r >= 5 for r in stock_returns):
        return "Low Risk: All stock returns are 5% or above."
    else:
        return "Medium Risk: All returns are positive, but at least one is below 5%."

def track_loan_repayment(initial_balance, monthly_payment):
    balance = initial_balance
    payments = []
    while balance > 0:
        balance -= monthly_payment
        balance = max(balance, 0)
        payments.append(balance)
    return payments

def track_stock_prices(stock_prices):
    tracked_prices = []
    for price in stock_prices:
        if price is None:
            continue
        if price >= 200:
            tracked_prices.append(price)
            break
        tracked_prices.append(price)
    return tracked_prices

def track_currency_exchange():
    exchange_rate = 290
    rates = []
    while exchange_rate <= 300:
        rates.append(exchange_rate)
        exchange_rate += 1
    return rates

def on_submit(_):
    returns = [float(val) for val in stock_returns_box.value.split(',')]
    result = classify_risk(returns)
    output_label.value = result

def on_loan_submit(_):
    initial_balance = float(initial_balance_box.value)
    monthly_payment = float(monthly_payment_box.value)
    payments = track_loan_repayment(initial_balance, monthly_payment)
    loan_output_label.value = "Remaining Balances: " + str(payments)

def on_stock_submit(_):
    stock_prices = [float(val) if val.lower() != 'none' else None for val in stock_prices_box.value.split(',')]
    tracked = track_stock_prices(stock_prices)
    stock_output_label.value = "Tracked Stock Prices: " + str(tracked)

def on_currency_submit(_):
    rates = track_currency_exchange()
    currency_output_label.value = "Exchange Rates: " + str(rates)

stock_returns_box = widgets.Text(description="Stock Returns", placeholder="Enter returns, e.g., 3, 7, -2")
submit_button = widgets.Button(description="Evaluate Risk")
output_label = widgets.Label()
submit_button.on_click(on_submit)

display(stock_returns_box, submit_button, output_label)

initial_balance_box = widgets.Text(description="Initial Loan", placeholder="Enter initial balance, e.g., 500000")
monthly_payment_box = widgets.Text(description="Monthly Payment", placeholder="Enter monthly payment, e.g., 25000")
loan_submit_button = widgets.Button(description="Track Loan Repayment")
loan_output_label = widgets.Label()
loan_submit_button.on_click(on_loan_submit)

display(initial_balance_box, monthly_payment_box, loan_submit_button, loan_output_label)

stock_prices_box = widgets.Text(description="Stock Prices", placeholder="Enter prices, e.g., 180, 190, None, 205")
stock_submit_button = widgets.Button(description="Track Stock Prices")
stock_output_label = widgets.Label()
stock_submit_button.on_click(on_stock_submit)

display(stock_prices_box, stock_submit_button, stock_output_label)

currency_submit_button = widgets.Button(description="Track Currency Exchange")
currency_output_label = widgets.Label()
currency_submit_button.on_click(on_currency_submit)

display(currency_submit_button, currency_output_label)
```
